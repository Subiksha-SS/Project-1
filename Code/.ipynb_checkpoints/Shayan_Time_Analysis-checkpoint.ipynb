```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import plotly.express as px
import scipy.stats as st
import numpy as py

```

```python
data_df = pd.read_csv('Resources/Crime_Data_from_2020_to_Present.csv')

main_df = data_df[['DR_NO', 'DATE OCC',
                   'TIME OCC', 'Vict Age',
                   'Vict Sex','Vict Descent',
                  'LAT', 'LON']]
main_df.count()
```

```python
# disclaimer for dropping missing age data
cleaning_df = main_df.set_index('DR_NO')
cleaning_df = cleaning_df.loc[(cleaning_df['Vict Age'] > 0) &
                              (cleaning_df['Vict Sex'] != "H") &
                              (cleaning_df['Vict Descent'] != "-")]
cleaning_df = cleaning_df.reset_index()
cleaning_df = cleaning_df.dropna(how='any')
```

```python
final_df = cleaning_df.rename(columns={'DR_NO': 'File Number',
                                      'DATE OCC': 'Date Occurred',
                                      'TIME OCC': 'Time Occurred',
                                      'LAT': 'Latitude',
                                      'LON': 'Longitude'})
final_df = final_df.set_index('File Number')
final_df.count()
final_df.head()
```

# Frequency vs Time for 2020-2023

```python
# binned_time = pd.cut(final_df['Time Occurred'], 4)
# print(binned_time.value_counts())
ax = final_df['Time Occurred'].plot.hist(density= True, bins = 4, xlim =[0,2359], color = 'c', alpha = 0.6)
final_df['Time Occurred'].plot.density(ax=ax)
tick_labels = ['Night', 'Morning', 'Afternoon', 'Evening']
tick_locations = [279.5, 838.5, 1397.5, 1956.5]
plt.xticks(tick_locations, tick_labels, rotation="horizontal")
plt.yticks([])
plt.ylabel("Frequency")
plt.title("Frequency of crimes in LA during different time frames for 2020-2023")
```

```python
#Separating the dd/mm/yyyy format into different columns (year, month, day) to have the year in a separate column for splitting the data into each year
final_df['Date Occurred'] = pd.to_datetime(final_df['Date Occurred']).dt.date
final_df['Time'] = pd.to_datetime(final_df['Date Occurred']).dt.time
final_df["day"] = final_df['Date Occurred'].map(lambda x: x.day)
final_df["month"] = final_df['Date Occurred'].map(lambda x: x.month)
final_df["year"] = final_df['Date Occurred'].map(lambda x: x.year)

# final_df.head()
```
